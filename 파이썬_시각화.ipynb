```python
import matplotlib.pyplot as plt  # 긴 이름을 plt 로 줄여서 사용

plt.plot([1,2,3,4])
#리스트의 값들을 y값이라고 가정하고 x는 자동으로 생성된다. 

plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,6], 'b-') # 블루의 실선으로 표현
plt.show()
```

```python
# x축과 y축에 이름 넣기
data_dict = {'data_x':[1,2,3,4,5], 'data_y':[2,4,5,10,8]}

plt.plot('data_x', 'data_y', data=data_dict)
plt.xlabel('X축')
plt.ylabel('Y축')

plt.show()
```

```python
# 라벨을 꾸미기

data_dict = {'data_x':[1,2,3,4,5], 'data_y':[2,4,5,10,8]}

plt.plot('data_x', 'data_y', data=data_dict)
plt.xlabel('X축', fontdict = {'color':'b', 'size':'xx-large'})
plt.ylabel('Y축',fontdict = {'color':'r', 'size':14 })             

plt.show()
```

```python
# 범례 lengend

data_dict = {'data_x':[1,2,3,4,5], 'data_y':[2,4,5,10,8]}

plt.plot('data_x', 'data_y', data=data_dict)
plt.xlabel('X축', fontdict = {'color':'b', 'size':'xx-large'})
plt.ylabel('Y축',fontdict = {'color':'r', 'size':14 })             
plt.legend()
plt.show()
```

```python
data_dict = {'data_x':[1,2,3,4,5], 'data_y':[2,4,5,10,8]}

plt.plot('data_x', 'data_y', data=data_dict, label = "Price ($)")
plt.xlabel('X축', fontdict = {'color':'b', 'size':'xx-large'})
plt.ylabel('Y축',fontdict = {'color':'r', 'size':14 })             
plt.legend()
plt.show()
```

```python
# 범례의 위치

data_dict = {'data_x':[1,2,3,4,5], 'data_y':[2,4,5,10,8]}

plt.plot('data_x', 'data_y', data=data_dict, label = "Price ($)")
plt.xlabel('X축', fontdict = {'color':'b', 'size':'xx-large'})
plt.ylabel('Y축',fontdict = {'color':'r', 'size':14 })             
plt.legend(loc = 8)   # loc 설정으로 그래프사엥 범례의 위치를 지정할 수 있다.
plt.show()            # 범례 위치 설정을 생략하면 자동으로 적당한 위치를 알아서 결정한다.
```

```python
data1 = {'data_x':[1,2,3,4,5], 'data_y':[2,4,5,10,8]}
data2 = {'data_x':[1,2,3,4,5], 'data_y':[3,5,9,7,7]}

plt.plot('data_x', 'data_y', data=data1, label = "Price ($)")
plt.plot('data_x', 'data_y', data=data2, label = "Demand (#)")

plt.xlabel('X축', fontdict = {'color':'b', 'size':'xx-large'})
plt.ylabel('Y축',fontdict = {'color':'r', 'size':14 })             
plt.legend(loc = 0, ncol=2, fontsize=12, shadow=True)     # 범례의 모양을 결정
plt.show()
```

```python
data1 = {'data_x':[1,2,3,4,5], 'data_y':[2,4,5,10,8]}
data2 = {'data_x':[1,2,3,4,5], 'data_y':[3,5,9,7,7]}

plt.plot('data_x', 'data_y', data=data1, label = "Price ($)")
plt.plot('data_x', 'data_y', data=data2, label = "Demand (#)")

plt.xlabel('X축', fontdict = {'color':'b', 'size':'xx-large'})
plt.ylabel('Y축',fontdict = {'color':'r', 'size':14 })             
plt.legend(loc = 0, ncol=2, fontsize=12, frameon=False)     # 범례의 모양을 결정
plt.show()
```

```python
# 산점도 scatter plot 분포를 나타내기에 적합

import numpy as np

np.random.seed(1)     # 항상 일정한 값으로 난수 발생

n = 50

a = (20 * np.random.rand(n)) ** 2
x = np.random.rand(n) # 0~1 난수 발생
y = np.random.rand(n)
colors = np.random.rand(n)

plt.scatter(x,y, c = colors, s = a)
plt.show()
```

```python
import numpy as np

np.random.seed(1)     # 항상 일정한 값으로 난수 발생

n = 50

a = (20 * np.random.rand(n)) ** 2
x = np.random.rand(n) # 0~1 난수 발생
y = np.random.rand(n)
colors = np.random.rand(n)

plt.scatter(x,y, c = colors, s = a, alpha = 0.4)
plt.show()
```

```python
import numpy as np

np.random.seed(1)     # 항상 일정한 값으로 난수 발생

n = 50

a = (30 * np.random.rand(n)) ** 2
x = np.random.rand(n)  # 0 ~ 1 사이로 난수 발생
y = np.random.rand(n)
z = np.random.rand(n)
colors = np.random.rand(n)


# plt.scatter(x,y, c = colors, s = a, alpha = 0.4)
fig = plt.figure(figsize = (10,10))  # 액자 설정
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x,y,z, s = a, c = colors, alpha = 0.4 )

plt.show()
```
